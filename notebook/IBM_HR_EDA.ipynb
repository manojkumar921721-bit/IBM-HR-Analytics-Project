```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/hr_attrition_cleaned.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
pd.crosstab(df["Department"], df["Attrition"])
```

```python
pd.crosstab(df["JobRole"], df["Attrition"], normalize="index") * 100
```

```python
sns.countplot(x="Attrition", data=df)
plt.title("Employee Attrition Distribution")
plt.show()
```

```python
sns.countplot(y="Department", hue="Attrition", data=df)
plt.title("Attrition by Department")
plt.show()
```

### Business Insight

- Sales department has the highest employee attrition.
- HR should prioritize retention programs for sales employees.
- Regular engagement surveys may help identify dissatisfaction early.

```python
plt.figure(figsize=(10,6))
sns.countplot(y="JobRole", hue="Attrition", data=df)
plt.title("Attrition by Job Role")
plt.show()
```

### Business Insight

- Sales Representatives and Laboratory Technicians have the highest attrition.
- These roles may require better career progression and compensation reviews.
